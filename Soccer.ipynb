# Scaling the Data

```python
from sklearn.model_selection import train_test_split
X_df = df_new.drop(columns=['Overall'])
y_df = df_new['Overall']
x_train, x_test, y_train, y_test = train_test_split(X_df,y_df)
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X_train = std.fit_transform(x_train)
X_test = std.transform(x_test)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
linearrig = model.score(X_test, y_test)
linearrig
```

```python
plt.bar(['linear'],[linearrig],label = 'results', color = 'red')
plt.ylabel('score')
plt.title('Linear Regression', size = 10)
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
fifa_rf = RandomForestRegressor(n_estimators=100)
fifa_rf.fit(X_train,y_train)
```

```python
fifa_rf.predict(X_test)
```

```python
randomforreg =fifa_rf.score(X_test, y_test)
randomforreg 
```

```python
plt.bar(['Random Forest'],[randomforreg],label = 'results', color = 'red')
plt.ylabel('score')
plt.title('Random Forest Regression', size = 10)
```

```python
df=df[['ID','Name','Age','Nationality','Overall','Potential','Position','Preferred Foot','International Reputation','Skill Moves','Jersey Number']].reset_index()
```

```python
df.describe
```

```python
df['total_score'] = (df.Overall/20 + df.Potential/20)
```

```python
df.total_score
```

```python
#Now, The Best Players are as follows
#df.Skill Moves = df.columns.str.replace(' ', '')
best = df.groupby('Skill Moves')['total_score'].agg(np.max).reset_index()
skill_scores = list(zip(best['Skill Moves'],best.total_score))
best_players = pd.DataFrame(columns=df.columns)
for index,i in enumerate(skill_scores):
    best_players = best_players.append(df[(df['Skill Moves'] == i[0] ) & (df.total_score == i[1])],ignore_index=True )
```

So, Finally based on average values of (Overall Rating, Potential Rating, International Reputation, Skill Moves)..... Below Players are the best among the Players of their Age.

```python
best_players.sort_values(by='Skill Moves',ascending=False)
```

Hence the above chart shows the best player of their particular skill group.

```python
best_players.sort_values(by='total_score',ascending=False)
```

Hence the above chart displays the top 6 player based on their total score and of different skill moves.
